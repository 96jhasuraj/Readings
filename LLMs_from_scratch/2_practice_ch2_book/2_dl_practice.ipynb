```python
import torch
from torch.utils.data import Dataset , DataLoader
```

```python
numbers = [i for i in range(102)]  
```

```python
class Data(Dataset):
    def __init__(self,number_ids,context_len,stride):
        self.X = []
        self.Y = []
        
        for i in range(0,len(number_ids)-context_len+1,stride):
            self.X.append(torch.tensor(number_ids[i:i+context_len]))
            self.Y.append(torch.tensor(number_ids[i+1:i+context_len+1]))
            #print(i)
    def __len__(self):
        return len(self.X)
    
    def __getitem__(self,idx):
        return self.X[idx],self.Y[idx]
```

```python
dataset = Data(numbers,4,4)
```

```python
len(dataset)
```

```python
def get_dataloader(txt,batch_size,context_len,stride,shuffle,drop_last):
    dataset = Data(txt,context_len,stride)

    dataloader = DataLoader(dataset,batch_size=batch_size,shuffle=shuffle,drop_last=drop_last)
    return dataloader
```
